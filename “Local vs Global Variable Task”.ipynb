### üü¢ ‚ÄúNested Functions with Scope (LEGB Rule)‚Äù

```python
#üß† LEGB ka full form:

#L = Local
#E = Enclosing
#G = Global
#B = Built-in
#LEGB = Python ka variable search rule hai
```

```python
#Q5.
x = 100   
def outer():
    x = 50  
    def inner():
        print(x)   
    inner()
outer()
print(x)
```

```python
#Q6.
x = 200   
def outer():
    x = 100 
    def inner():
        x = 50
        print(x)   
    inner()
outer()
print(x)
```
